```python
#Lets import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Lets import th dataset by creating a varioble called "epl_df"
epl_df = pd.read_csv("EPL_20_21.csv")
epl_df.head()
```

```python
#lets check values by using ".info()" of the pandas library
epl_df.info()
```

```python
epl_df.describe()
#Fact ".describe()" function works on numeric values 
```

```python
epl_df.isna().sum()
#Count of nulls in the dataset by the function ".isna().sum()"
```

```python
#We need create other two extra columns in the dataset: 1.Minutes per match & 2.Goals per match
epl_df['MinutesPerMatch'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']= (epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df.head()
```

```python
#Lets Find the Total Goals in EPL 2020-2021 season
#Create a new variable as tortalgoals and using pandas ".sum()" function.
Totalgoals = epl_df['Goals'].sum()
Totalgoals
```

```python
#Also find "Total Penalty Goals" in EPL 2020-2021 season.
Total_penalty_goals = epl_df['Penalty_Goals'].sum()
Total_penalty_goals
```

```python
#Lets Check penlaty attempts in EPL 2020-2021 season.
Total_penalty_attempts = epl_df['Penalty_Attempted'].sum()
Total_penalty_attempts
```

```python
#Creating a pie Chart for Penalties vs Scored
#Creating a new variable pl_not_scored by subtracting Penalty_Attempted & Total_penlaty_goals
#I am using "autopct = '%0f%%'" to display the % values in the pie chart.
#I mentioned my parameters like data,labels,color.
plt.figure(figsize=(14,7))
pl_not_scored = epl_df['Penalty_Attempted'].sum() - Total_penalty_goals
data =[pl_not_scored, Total_penalty_goals]
labels = ['Penlaties Missed','Penlaties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors = color ,autopct = '%.0f%%')
plt.show()
```

```python
#Total no:of unique positions in the football
#Use .unique() function of numpy to get the different types positions in the game.
epl_df['Position'].unique()
```

```python
#Lets find total FW(forward Players) in the dataset
epl_df[epl_df['Position']=='FW']
#in the output we can see , there are 80 rows which means , there are total 80 forward players in the EPL season 2020-2021
```

```python
#Lets the Find the Players form the Different nations.
#Here I used numpy function ".size()" and also ".unique()" to get the count of total nations. 
np.size((epl_df['Nationality'].unique()))

#So there are total 59 nations inth season 2020 -2021(EPL)
```

```python
#Most players from whih countries
#Lets create a new Variable.
nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind ='bar',figsize=(13,6),color = sns.color_palette("magma"))

#we could see Top 10 nations , from which most players come.
```

```python
#Clubs with the max players in the squad
#Lets counts the value of the Club column of the dataset by using ".nlarget()" function to get Top 5 clubs.
epl_df['Club'].value_counts().nlargest(10).plot(kind ='bar',figsize=(14,7),color = sns.color_palette("viridis"))
```

```python
#Clubs with the least players in the squad
epl_df['Club'].value_counts().nsmallest(10).plot(kind ='bar',figsize=(14,7),color = sns.color_palette("viridis"))
```

```python
#Players based their age group.
#Firsly create some variables according to the age.
under20 = epl_df[epl_df['Age']<= 20]
age20_25 = epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age25_30 = epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
Above30 = epl_df[epl_df['Age']>30]
```

```python
#Create a array for the above variables by using .array() function
x = np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
labels = ["<=20",">20 &<=25",">25 & <=30",">30"]
plt.title("Total Payers with Age Groups", fontsize =20)
plt.pie(x, labels = labels ,colors = color, autopct = "%.1f%%")
plt.show()
```

```python
#Total <20 players in each club
players_under20 = epl_df[epl_df['Age']<20]
players_under20['Club'].value_counts().plot(kind ='bar',figsize=(14,7), color = sns.color_palette("viridis"))
```

```python
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x ='Club',y ='Goals', data = Goals_by_clubs.sort_values(by="Goals"),palette='rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=30)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(30,10)
plt.title('Plot of Clubs vs Total Goals', fontsize = 20)
```
