```python
#lets import important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('play store.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
#Lets find the count of app per genres
app_count = df['Genres'].value_counts().sort_values(ascending=False).head(50)
app_count.reset_index()
```

```python
app_count.plot(kind='bar',figsize=(16,8))
plt.xlabel('Genre Type',fontsize=13)
plt.ylabel('Count of Apps',fontsize=13)
plt.title('App Count per Genre',fontsize=20)
```

```python
app_count = df['Genres'].value_counts().sort_values(ascending=False).head(20)
app_count.reset_index()
```

```python

plt.figure(figsize=(15,15))
plt.title("% of App for Top 20 Genre",fontsize=20)
plt.pie(app_count.values,labels=app_count.index,labeldistance=1.2,autopct='%1.2f%%',startangle=60)
plt.show()
```

```python
#Create a new variable of app list with only 5 rating. 
five_star = df[df['Rating']==5]
five_star
```

```python
genre_rate = five_star.groupby(by='App').max()[['Rating','Genres']].sort_values(by='Rating',ascending =False)
genre_rate
```

```python
#Genres with 5 Star rating counts
genres_five_star = five_star['Genres'].value_counts()
genres_five_star.reset_index()
```

```python
#Genres with HIghest 5 Star Ratings
genres_five_star.plot(kind='bar',figsize=(16,8),color='green')
plt.title('Genres With 5 Star Rating Count', fontsize=20)
```

```python
#Similarly another variable , to filter the list of apps with highest installs or Downloads
high_review = df[df['Reviews']>10000].sort_values(by='Reviews',ascending=False)
high_review
```

```python
# List of App & its rating,& Downloads, which has Highest reviews count
highest_review = df.groupby(by='App').max()[['Reviews','Rating']].sort_values(by='Reviews',ascending=False).head(50)
highest_review.reset_index()
```

```python
#presentation of apps with Highest Reviews gained
highest_review.plot(kind='bar',color='blue',figsize=(18,9))
plt.title('Apps With Highest Reviews Count',fontsize=20)
plt.ylabel('Reviews Count in Millions',fontsize=13)
plt.xlabel('App Name',fontsize=13)
```

```python
#Apps with Highest downloads 
highest_installs = df.groupby(by='App').max()['installs'].sort_values(ascending=False).head(50)
highest_installs.reset_index()
```

```python
# App with highest downloads
highest_installs.plot(kind='bar',x='App',y='installs',figsize=(18,6))
plt.title('Highest Downloaded Apps',fontsize=20)
plt.ylabel('Downloaded in Billions',fontsize=14)
plt.xlabel('App Name',fontsize=14)
```

```python
# Year of highest app updates received
year_update = df['year of update'].value_counts()
year_update.reset_index()
```

```python
#Presentation of above data
year_update.plot(kind='bar',figsize=(18,9),color='red')
plt.title('Year with Highest App Upates',fontsize=20)
```
