```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("disney_plus_titles.csv")
df.head()
```

```python
types = df['type'].value_counts()
types.reset_index()
```

```python
plt.figure(figsize=(10,10))
plt.title("% Of Total Movies & TV-Show",fontsize=20)
plt.pie(types.values,labels=types.index,labeldistance=1.2,autopct='%1.2f%%',startangle=90)
```

```python
country = df['country'].value_counts()
country.reset_index()
```

```python
country.head(15).plot(kind ='bar',figsize=(13,6),color = sns.color_palette("magma"))
plt.xticks(rotation=90)
```

```python
top_10_genres = df['genre'].value_counts().head(30)
top_10_genres.reset_index()
```

```python
top_10_genres.plot(kind='bar', figsize=(20,10), color = sns.color_palette("viridis"))
plt.title("Top Genres With Movie Count", fontsize=20,color= 'blue')
```

```python
# Top Shows in dataset
shows = df[df['type']=='TV Show']
shows
```

```python
# Top Genres in Tv Shows
shows_genres = shows['genre'].value_counts()
shows_genres.reset_index().head(15)
```

```python
#Count of Tv shows
release_year = shows['release_year'].value_counts()
release_year.reset_index().head(20)
```

```python
release_year.plot(kind = 'bar', figsize=(16,8), color='green')
plt.title('Tv-Shows Counts', fontsize=20, color = 'blue')
```

```python
# Lets create a new variable for Tvshows which are released in 2021
tv_shows = df[(df['type']=='TV Show') & (df['release_year']==2021)]
tv_shows
```

```python
#Latest Shows streaming on Disney+ (2021 released)
shows_2021 = tv_shows.groupby('title').max()[['release_year','rating','genre','duration']].sort_values(by='release_year' ,ascending=False)
shows_2021
```

```python
rating = df['rating'].value_counts().reset_index()
rating
```

```python
plt.figure(figsize=(18,9))
plt.title("Total Tv Shows & Movies Per Rating",fontsize=20,color='Green')
ax = sns.barplot(data=rating, x="index", y='rating',linewidth=2,edgecolor='black')
plt.xlabel("Rating Type",fontsize=13,color='black')
plt.ylabel("Count",fontsize=13,color='black')
```
