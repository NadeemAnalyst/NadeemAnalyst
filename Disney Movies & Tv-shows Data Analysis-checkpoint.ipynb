```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('disney 2019-2020.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['added_at'] = pd.to_datetime(df['added_at'])
df['released_at'] = pd.to_datetime(df['released_at'])
```

```python
df.shape
```

```python
types = df['type'].value_counts()
types
```

```python
plt.figure(figsize=(18,9))
plt.title("Movies & Tv-shows", fontsize=20, color='blue')
plt.pie(types.values,labels=types.index,labeldistance=1.2,autopct='%1.2f%%',startangle=90)
```

```python
df = df.rename(columns={'Column1':'rating'})
```

```python
df['rating']=df['rating'].replace('PASSED','Passed')
```

```python
rating = df['rating'].value_counts().sort_values(ascending=False).reset_index()
rating
```

```python
plt.figure(figsize=(18,9))
plt.title("Count Of Movies And Tv-Shows Per Rating",fontsize=20,color='green')
ax=sns.barplot(data=rating, x='index',y='rating',linewidth=2,edgecolor='black')
plt.ylabel("Count",fontsize=14,color='green')
plt.xlabel("Type Of Rating",fontsize=14,color='green')
```

```python
movies = df[df['type']=='movie']
movies
```

```python
series = df[df['type']=='series']
series
```

```python
movies_2019 = movies[(movies.added_at >= '01-01-2019')&(movies.added_at <= '31-12-2019')]
movies_2019
```

```python
#Top Movies in 2019, with highest imdb rating
top_movies_2019 = movies_2019.groupby('title').max()[['imdb_rating','released_at','added_at','awards','metascore']].sort_values(by='imdb_rating',ascending=False).head(25)
top_movies_2019
```

```python
#Top Movies in 2020, with highest imdb rating
movies_2020 = movies[(movies.added_at >= '01-01-2020')&(movies.added_at <= '31-12-2020')]
movies_2020
```

```python
#Top Movies in 2020, with highest imdb rating
top_movies_2020 = movies_2020.groupby('title').max()[['imdb_rating','added_at','released_at','awards','metascore']].sort_values(by ='imdb_rating', ascending =False).head(25)
top_movies_2020
```

```python
#Movies with >1Million Imdb_votes
top_votes = movies[movies['imdb_votes']>=1000000].sort_values(by='imdb_votes',ascending=False)
top_votes
```

```python
#Movies with >1 Million Imdb_votes
top_votes.plot(kind='bar',x='title',y='imdb_votes',figsize=(16,8),color='green',edgecolor='black',linewidth=2)
plt.title("Top Movies With 1 Million Imdb Votes",fontsize=20,color='blue')
plt.ylabel("Votes In Millions",fontsize=14,color='blue')
plt.xlabel("Movies Name",fontsize=14,color='blue')
```
