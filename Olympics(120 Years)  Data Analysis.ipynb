```python
#Lets import all the important libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data set by using pandas function ".read_csv()"
athelets = pd.read_csv('athlete_events.csv')
regions = pd.read_csv('noc_regions.csv')

```

```python
#Lets check the Rows of the Dataset
athelets.head()
```

```python
regions.head()
```

```python
#Merge the datasets by using pandas function {.merge(xxxx, how = 'left', on = 'NOC')}
#Here i am joining both the datasets from Left, and my common column is 'NOC' which is present inthe both datasets.
athelets_df = athelets.merge(regions , how = 'left', on = 'NOC')
athelets_df.head()
```

```python
#Look the output , we have added 2 extra columns (region & notes)

#Lets make the column names consistent\
#use pandas function ".rename(column {'xxxxx':'yyyy','aaaaa':'bbbb'}, inplace = True);"
#inplace = True, makes the changes in dataset permanent. 
athelets_df.rename(columns ={'region':'Region','notes':'Notes'},inplace = True);
athelets_df.head()
```

```python
#print the info of the dataset by ".info()"
athelets_df.info()
```

```python
#to display some statistical values of the dataset by using ".describe()"
athelets_df.describe()
```

```python
#If you observe the Result , The min value shows the 1896 in which the Olympics started.recent on eis 2016


#I create a new variable.check the null values in the dataset , by using ".isna()".
nan_values = athelets_df.isna()
#another variable to get result of nulls in Boolian values by using ".any()"
nan_columns = nan_values.any()
nan_columns

```

```python
#See the reult, we have 6 columns , which are having nulls or missing values.

#Lets see the Total Count of nulls by ".isnull().sum()"
athelets_df.isnull().sum()
```

```python
#Lets check the India Details
athelets_df.query('Team == "India"').head(10)
```

```python
#lets check the japan details
athelets_df.query('Team == "Japan"').head(10)
```

```python
#top 10 countries participated in Olympics
top_10countries = athelets_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10countries
```

```python
#Plot the result in graph like Barplot
plt.figure(figsize=(14,7))
#plt.xticks(rotation=20)
plt.title("Top 10 Countries participating in Olympics")
sns.barplot(x = top_10countries.index , y = top_10countries, palette = 'Set1');
```

```python
#Age distribution of the participants
plt.figure(figsize=(12,6))
plt.title("Age Distribution of the Athelets in Oplympics")
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athelets_df.Age, bins = np.arange(10,80,2), color ='orange', edgecolor = 'black')
#I used numpy function np.arange(10,80,2) , means from 10 to 80 with the gap 2.
```

```python
#Lets make a list of winter sports in the Olympics
#i used.Season wjich is a columns in the dataset and == 'Winter' which is a sport ".Sport"
winter_sports = athelets_df[athelets_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#lets do the same of above , but now with summer sports
summer_sports = athelets_df[athelets_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
#lets check the Participation of Male & Female in Olympics from 1896 to 2016
gender_counts = athelets_df.Sex.value_counts()
gender_counts
#look i used ".value_counts()" with column of Sex , to get the total count of both genders.
```

```python
#I make a Bar Plot of male and female participants
plt.figure(figsize=(14,7))
plt.title('Male and Female Particants in Olympics')
sns.barplot(x = gender_counts.index, y =gender_counts , palette = 'Set1')
```

```python
#Total number of Medals
athelets_df.Medal.value_counts()
```

```python
#total participants of Femlaes in all seasons of olympics
women_olympics = athelets_df[(athelets_df.Sex == 'F')&(athelets_df.Season =='Summer')]
```

```python
#Create a Count plot of the women_olympics
#I used seaborn sunction style='darkgrid' with ".countplot()"
sns.set(style ="darkgrid")
plt.figure(figsize=(14,7))
plt.title("Women Participation on Olympics")
sns.countplot(x = 'Year', data = women_olympics, palette = "Spectral")
```

```python
#Lets check the athelets with Gold Medals
goldMedals = athelets_df[(athelets_df.Medal =='Gold')]
goldMedals.head()
```

```python
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
#Gold Medalists beyond 60 years old
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
#Lets see the sporting events of 60 years athelets who have won Gold Medals
sport_event = goldMedals['Sport'][goldMedals['Age']>60]
sport_event
```

```python
#I Create a count Plot of the above sport events
#Here I used seaborn function ".countplot(variable_name)"
plt.figure(figsize=(14,7))
plt.tight_layout()
plt.title('Gold Medals for athelets above 60 years')
sns.countplot(sport_event)
```

```python
#Gold Medals for each Country
#I used pandas function ".value_counts()" to get count of Gold Medals & ".reset_index(name = "Medal")"
#.head(5) to get top 5 Gold Medals winners
goldMedals.Region.value_counts().reset_index(name='Medal').head(10)
```

```python
#I create a cat plot of the Top 5 countries with Gold Medals
Totalgoldmedals =goldMedals.Region.value_counts().reset_index(name='Medal').head(5)
#create a new variable "G"
g = sns.catplot(x = "index", y ="Medal",data = Totalgoldmedals, height = 5, kind = "bar", palette = "rocket")
g.despine(left =True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country')
```

```python
#Recent Rio Olympics data
#max_year is my new variable to find the max no:of medals in the year of 2016
max_year = athelets_df.Year.max()
print(max_year)

#My another variable "team_names"
team_names = athelets_df[(athelets_df.Year== max_year)&(athelets_df.Medal =="Gold")].Team
team_names.value_counts().head(10)
```

```python
#Using the above result, i carete a barplot
sns.barplot(x = team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel('Countries Medals for the year of 2016')
```
