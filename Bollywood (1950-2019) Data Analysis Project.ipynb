```python
#Lets import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load The dataset by pandas function
df = pd.read_csv('50 Years Bollywood.csv')
df.head()
```

```python
#check the null values
df.isnull().sum()
```

```python
#statistical values of the dataset
df.describe()
```

```python
# No:of Rows & columns 
df.shape
```

```python
#Total count of movies from 1950-2019 recent times
count_movies = df['release year'].value_counts().reset_index()
count_movies
```

```python
count_movies.plot(kind='bar',figsize=(20,10))
```

```python
# Movies count per genre wise
genre = df['genres'].value_counts().sort_values(ascending=False).head(10)
genre.reset_index()
```

```python
plt.figure(figsize=(20,10))
plt.pie(genre.values,labels=genre.index, labeldistance=1.2,autopct='%1.2f%%',shadow=True,startangle=60)
plt.title('Top 10 Genres With Movies Count', fontsize =15,color='orange')
plt.plot()
```

```python
#movies with highest imdb votes
votes = df[df['imdb_votes']>70000]
votes.reset_index()
```

```python
votes.plot(kind='bar',x='title',y='imdb_votes',figsize=(20,10),color='green')
plt.title('Movies With Highest Imdb Votes (>70000)',fontsize=25)
plt.xlabel('Movie Name',fontsize=18)
plt.ylabel('Imdb Votes',fontsize=18)
plt.xticks(rotation=45,fontsize=10)
```

```python
# movies highest imdb rating
rates = df[df['imdb_rating']>8.5].value_counts()
rates.reset_index()
```

```python
# group of data on the base of rating >8.5
df.groupby(by = 'title').max()[['imdb_rating','imdb_votes','genres']].sort_values(by=['imdb_rating'],ascending=False).head(30).reset_index()
```

```python
# movies with least rates & votes
df.groupby(by = 'title').max()[['imdb_rating','imdb_votes','genres']].sort_values(by=['imdb_rating'],ascending=True).head(30).reset_index()
```

```python
# list of old movies from 1950-1999 
movie_1950 = df[df['release year']<=1999]
movie_1950
```

```python
# top rated old movies
a = movie_1950.groupby(by='title').max()[['imdb_rating','imdb_votes']].sort_values(by='imdb_rating',ascending=False).head(15).reset_index()
a
```

```python
a.plot(kind='bar',x='title',y='imdb_rating',figsize=(20,10),color='yellow')
plt.title('Top Movies Of (1950-1999) With Imdb Rating',fontsize=25,color='green')
plt.xlabel('Movie Name',fontsize=18)
plt.ylabel('Imdb rating',fontsize=18)
plt.xticks(rotation=45,fontsize=10)
```

```python
# movies with highest running time 
b = df.groupby(by = 'title').max()['runtime'].sort_values(ascending=False).head(10).reset_index()
b
```

```python
b.plot(kind='bar',x='title',y='runtime',figsize=(20,10),color='red')
plt.title('Top 10 Movies Highest Runtime',fontsize=25,color='blue')
plt.xlabel('Movie Name',fontsize=18)
plt.ylabel('Run Time In Minutes',fontsize=18)
plt.xticks(rotation=45,fontsize=10)
```

```python
#movies with least running time
c =df.groupby(by = 'title').max()['runtime'].sort_values(ascending=True).head(10).reset_index()
c
```

```python
c.plot(kind='bar',x='title',y='runtime',figsize=(20,10),color='pink')
plt.title('Movies With Least Runtime',fontsize=25,color='orange')
plt.xlabel('Movie Name',fontsize=18)
plt.ylabel('Run Time in Minutes',fontsize=18)
plt.xticks(rotation=45,fontsize=10)
```
