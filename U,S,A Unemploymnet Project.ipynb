```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Create a new varibale called "unepm"
unemp = pd.read_csv('unemployment_rate_data.csv')
```

```python
unemp['date'] = pd.to_datetime(unemp['date'])
```

```python
unemp.head()
unemp.describe()
```

```python
#lets plot the line plot date ve unrate
plt.figure(figsize=(16,8))
plt.plot(unemp['date'],unemp['unrate'])
plt.title("Unemployment rate of US from 1948 to 2021", size = 25)
```

```python
#create presentation for unemployment rate of women
plt.figure(figsize=(18,6))
sns.lineplot(unemp['date'], unemp['unrate_women'], color ='red')
plt.title("Unemployment rate of women from 1948-2021", size = 25)
```

```python
#similar presentation about men
plt.figure(figsize=(18,6))
sns.lineplot(unemp['date'], unemp['unrate_men'], color ='green')
plt.title("Unepmloyment rate of men from 1948-2021", size = 25)
```

```python
# unemployment rate from 1980 t0 1990
plt.figure(figsize=(18,6))
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1990')], x= 'date', y ='unrate')
plt.title("unemployment rate of citizens(US) from 1980 to 1990", size = 25)
```

```python
#unemployment rate of women from 1980 to 1990
plt.figure(figsize=(18,6))
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1990')], x= 'date', y ='unrate_women', color = 'red')
plt.title("unemployment rate of women from 1980 to 1990", size = 25)
```

```python
#Lets the Unemployment of men from 2019 - 2021(During covid period)
plt.figure(figsize=(18,6))
sns.lineplot(data =unemp[(unemp.date >= '1/1/2019')&(unemp.date<= '12/1/2021')], x= 'date', y ='unrate_men', color = 'violet')
sns.lineplot(data =unemp[(unemp.date >= '1/1/2019')&(unemp.date<= '12/1/2021')], x= 'date', y ='unrate_women', color = 'orange')
plt.title("unemployment rate of men & women from 2019-2021(During covid19)", size = 25)
plt.xlabel("Year 2019-2021")
plt.ylabel("unrate")
```

```python
plt.figure(figsize=(18,6))
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1984')], x= "date", y ="unrate_16_to_17",label ="unrate_16_to_17", color = 'red')
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1984')], x= "date", y ="unrate_18_to_19",label ="unrate_18_to_19", color = 'orange')
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1984')], x= "date", y ="unrate_20_to_24",label ="unrate_20_to_24", color = 'red')
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1984')], x= "date", y ="unrate_25_to_34",label ="unrate_25_to_34", color = 'violet')
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1984')], x= "date", y ="unrate_35_to_44",label ="unrate_35_to_44", color = 'pink')
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1984')], x= "date", y ="unrate_45_to_54",label ="unrate_45_to_54", color = 'violet')
sns.lineplot(data =unemp[(unemp.date >= '1/1/1980')&(unemp.date<= '12/1/1984')], x= "date", y ="unrate_55_over",label ="unrate_55_over", color = 'green')
plt.xlabel("unrate")
plt.ylabel("Time")
plt.title("Unemployment rate of all ages during 1980's Economic crisis" , size=25)
```

```python
#lets do the same in covid 19 period
plt.figure(figsize=(18,6))
sns.lineplot(data =unemp[(unemp.date >= '1/1/2020')&(unemp.date<= '12/1/2020')], x= "date", y ="unrate_16_to_17",label ="unrate_16_to_17", color = 'red')
sns.lineplot(data =unemp[(unemp.date >= '1/1/2020')&(unemp.date<= '12/1/2020')], x= "date", y ="unrate_18_to_19",label ="unrate_18_to_19", color = 'orange')
sns.lineplot(data =unemp[(unemp.date >= '1/1/2020')&(unemp.date<= '12/1/2020')], x= "date", y ="unrate_20_to_24",label ="unrate_20_to_24", color = 'red')
sns.lineplot(data =unemp[(unemp.date >= '1/1/2020')&(unemp.date<= '12/1/2020')], x= "date", y ="unrate_25_to_34",label ="unrate_25_to_34", color = 'violet')
sns.lineplot(data =unemp[(unemp.date >= '1/1/2020')&(unemp.date<= '12/1/2020')], x= "date", y ="unrate_35_to_44",label ="unrate_35_to_44", color = 'pink')
sns.lineplot(data =unemp[(unemp.date >= '1/1/2020')&(unemp.date<= '12/1/2020')], x= "date", y ="unrate_45_to_54",label ="unrate_45_to_54", color = 'violet')
sns.lineplot(data =unemp[(unemp.date >= '1/1/2020')&(unemp.date<= '12/1/2020')], x= "date", y ="unrate_55_over",label ="unrate_55_over", color = 'green')
plt.ylabel("unrate")
plt.xlabel("period")
plt.title("Unemployment rate of all ages during covid_19 period" , size=25)
```
