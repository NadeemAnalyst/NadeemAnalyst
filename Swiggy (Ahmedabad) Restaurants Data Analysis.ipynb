```python
#lets import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
#load the dataset
df = pd.read_csv('Swiggy_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# crate a variable for anmedabad city.Bring the data of ahmedabad from the whole dataset.
ahmd = df[df['city']=='Ahmedabad']
ahmd
```

```python
#count of restaurants in the city
ahmd['name'].shape
```

```python
#creating another variable , which is about the restaurants >4.6 Avg Ratings
ahmd_res = ahmd[ahmd['avgRating']>4.6]
ahmd_res
```

```python
#Top Rated restaurants in Ahmedabad city with >4.6 rating
plt.figure(figsize=(14,7))
sns.barplot(x ='name', y ='avgRating', data = ahmd_res)
plt.xticks(rotation = 45, fontsize=15,color='violet')
plt.title("Top Rated Restaurants in Ahmedabad city with >4.6 rating", fontsize= 20,color='purple')
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(x ='cuisines', y ='avgRating', data = ahmd_res, color='orange')
plt.xticks(rotation = 45, fontsize=15,color='green')
plt.title("Top Rated Cuisines In Ahmedabad", fontsize= 20,color='green')
```

```python
#Restaurants with least ratings
ahmd_res2 = ahmd[ahmd['avgRating']<=2.9]
ahmd_res2
```

```python
#creating a pivot table for the least rated restaurants
a = pd.pivot_table(ahmd_res2,'avgRating',index='name')
result = a.sort_values('avgRating',ascending=False)
result
```

```python
result.plot(kind='bar', figsize=(14,7),color='black')
plt.xticks(rotation=45, fontsize=13,color='green')
plt.title("Restaurants In Ahmedabad With <3 Avg Rating", fontsize=20, color='green')
```

```python
#Lets find the list of restaurants in Ahmedbad which delivers food in <60 min
b = pd.pivot_table(ahmd_res, 'deliveryTime', index=['name'])
Result = b.sort_values('deliveryTime', ascending=True)
Result
```
