```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Hulu Dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
year = df[df['release_year']>2020]
year
```

```python
#Count Of Tv Shows & Movies As Per Genre
count = df['listed_in'].value_counts().head(10)
count.reset_index()
```

```python
count.plot(kind='bar',color='yellow',figsize=(18,6))
plt.ylabel("Count", fontsize=13,color='red')
plt.xlabel("Genre",color='red',fontsize=13)
plt.title("Top 10 Genres With Tv Shows & Movie Count",fontsize=20,color='green')
plt.xticks(rotation=45)
```

```python
#movies from different nations
country = df['country'].value_counts().sort_values(ascending=False).head(10)
country.reset_index()
```

```python
country.plot(kind='bar',color='blue',figsize=(16,8))
plt.xlabel("Count Of Movies ", fontsize=13,color='green')
plt.ylabel("Country",color='green',fontsize=13)
plt.title("Top 10 Genres With Movie Count",fontsize=20,color='red')
plt.xticks(rotation=45)
```

```python
rate = df['rating'].value_counts().sort_values(ascending =False).head(9)
rate.reset_index()
```

```python
rate.plot(kind='line',color='black',figsize=(16,8))
plt.xlabel("Count Of Movies ", fontsize=13,color='red')
plt.ylabel("Rateing",color='red',fontsize=13)
plt.title("Ratings With Movie Count",fontsize=20,color='blue')
```

```python
#Count of Total Movies & Total TV Shows
Type = df['type'].value_counts()
Type.reset_index()
```

```python
#Movies vs Tv shows
plt.figure(figsize=(16,8))
plt.pie(Type.values,labels=Type.index, labeldistance=1.2,autopct='%1.2f%%', shadow=True,startangle=60)
plt.title('Movies vs Tv Shows', fontsize =15)
plt.plot()
```
